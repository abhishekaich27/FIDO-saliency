Flipping the top pixels and see how much logodds drop in terms of the top-1 classification error.

First, we load the logodds from 4 folders, and take the common parts out.
- AE_p_b
- AE_vbd
- VAE_p_b (5 samples)
- VAE_vbd

```python
%matplotlib inline
import torch
import os
import pandas as pd
import seaborn as sns
```

Take out all the files end with 'records.th'

```python
def get_file_names(directory):
    directory = os.path.join('../imgs', directory)
    result = []
    for filename in os.listdir(directory):
        if filename.endswith("records.th"): 
            result.append(filename)
    return result
```

```python
arr = []
identifiers = ['0908_ae_p_b', '0910_ae_vbd_final_0.1', '0908_vae_p_b', '0910_vae_vbd_final_0.1']
for directory in identifiers:
    filenames = get_file_names(directory)
    arr.append(filenames)

# Find the smallest sets
smallest = 201
small_idx = ''
for i, filenames in enumerate(arr):
    if len(filenames) < smallest:
        smallest = len(filenames)
        small_idx = i
print small_identifer
```

```python
import seaborn as sns
import pandas as pd
import numpy as np

IMG_SIZE = 224 * 224

def prepare_pd_table():
    result = []
    for i in xrange(len(identifiers)):
        identifier = identifiers[i]
        print identifier
        
        for j in xrange(len(arr[i])):
            orig_log_odds, all_log_odds, unnormalized_img, imp_vector = \
                torch.load(os.path.join('../imgs', identifier, arr[i][j]))
            
            if isinstance(all_log_odds, dict):
                for key in all_log_odds:
                    log_odds_drop = orig_log_odds - all_log_odds[key]
                    result.append([identifier, j, key, log_odds_drop])
            else:
                percs = [0.005, 0.01, 0.02, 0.05, 0.09]
                num_flippings = [int(perc * IMG_SIZE) for perc in percs]
                for idx, k in enumerate(num_flippings):
                    log_odds_drop = orig_log_odds - all_log_odds[k]
                    result.append([identifier, j, percs[idx], log_odds_drop])
                
    result = pd.DataFrame(result)
    result.columns = ['method', 'img_index', 'num_flippings', 'odds_diff']
    return result
```

```python
table = prepare_pd_table()

import seaborn as sns
ax = sns.boxplot(x="num_flippings", y="odds_diff", hue="method", data=table)
```

```python
IMG_SIZE = 224 * 224
table = prepare_pd_table([100, 1000, 2000, 4000])

ax = sns.boxplot(x="num_flippings", y="odds_diff", hue="method", data=table)
```

```python
identifiers = ['0912_ae_vbd_0.1', '0912_ae_vbd_1.0', '0912_ae_vbd_10.0', '0912_ae_p_b']
arr = [get_file_names('../imgs/{}'.format(i)) for i in identifiers]

def prepare_pd_table(arr, identifiers):
    result = []
    for i in xrange(len(identifiers)):
        identifier = identifiers[i]
        
        for j in xrange(len(arr[i])):
            orig_log_odds, all_log_odds_dict, unnormalized_img, imp_vector = \
                torch.load(os.path.join('../imgs', '%s' % identifier, arr[i][j]))
            
            for key in all_log_odds_dict:
                log_odds_drop = orig_log_odds - all_log_odds_dict[key]
                result.append([identifier + '(n = %d)' % (len(arr[i])), j, key, log_odds_drop])
    result = pd.DataFrame(result)
    result.columns = ['method', 'img_index', 'num_flippings', 'odds_diff']
    return result
```

```python
table = prepare_pd_table(arr, identifiers)
ax = sns.boxplot(x="num_flippings", y="odds_diff", hue="method", data=table)
```
